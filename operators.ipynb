# Python Operatoren / *Operators*

Python-Operatoren sind spezielle Symbole, mit denen bestimmte Operationen auf einem oder mehreren Operanden ausgeführt werden. Variablen, Werte oder Ausdrücke können dabei als Operanden verwenden werden. Zum Beispiel wird der Additionsoperator (+) in Python verwendet, um Additionen mit zwei Variablen, Werten oder Ausdrücken durchzuführen.

---

*Python operators are special symbols used to perform specific operations on one or more operands. The variables, values, or expressions can be used as operands. For example, Python's addition operator (+) is used to perform addition operations on two variables, values, or expressions.*

# Arithmetische Operatoren / *Arithmetic Operators*

Arithmetische Operatoren in Python werden verwendet, um grundlegende mathematische Operationen wie Addition, Subtraktion, Multiplikation usw. durchzuführen.

---

*Python Arithmetic operators are used to perform basic mathematical operations such as addition, subtraction, multiplication, etc.*

````python
a = 10
b = 3
````

| Operator | Meaning             | Example  | Result      |
| -------- | ------------------- | -------- | ----------- |
| `+`      | Addition            | `a + b`  | `13`        |
| `-`      | Subtraction         | `a - b`  | `7`         |

#### Integer Division erzeugt Fließkommazahlen / *Integer Division Produces Float*

In Python kann die Division den Datentyp des Ergebnisses verändern. Der Operator / gibt immer eine Fließkommazahl (float) zurück – selbst dann, wenn beide Zahlen ganze Zahlen (Integer) sind.

---

In Python, division can change the data type of the result. The `/` operator returns a float, even if both numbers are integers, the / operator returns a float.

````python
x = 7
y = 2

result = x / y

print(result)
print(type(result))
````

````python
# a and b are integers (int)
a = 8
b = 4

# When using /, Python performs true division, so the result becomes a float 5.0 instead of an integer 5.
result = a / b
print(type(result))
````

### Beibehalten von Integer Resultaten / *Keeping an Integer Result*

Wenn gewünscht ist, dass das Resultat ein integer bleibt, kann Ganzzahldivision (Floor-Division) `//` benutzt werden.

---

*If you want the result to stay an integer, use floor division `//`.*

````python
a = 10
b = 2

result = a // b

print(result)
print(type(result))
````

Welches Ergebnis kommt heraus, wenn -7 durch 2 geteilt wird?

---

*Floor division `//` does not just remove decimals. It rounds down to the nearest integer.
What do you think would be the result if we divide `-7` by `2`?*

````python
print(-7 // 2)
````

Warum?

`-7 / 2 = -3.5`

Die Floor-Division geht zur nächstkleineren ganzen Zahl, also zu -4 und nicht zu -3.

---

*Why?*

`-7 / 2 = -3.5`

*Floor division goes to the next lower integer, which is `-4`, not `-3`.*

# Logische Operatoren / *Logical Operators*

Alle anderen Werte werden als True behandelt.

Es gibt drei Logikoperatoren in Python: "and", "or" und "not". Sie müssen in Kleinbuchstaben geschrieben werden.

*Python logical operators are used to form compound Boolean expressions. Each operand for these logical operators is itself a Boolean expression. All other values are treated as True.*

*There are three logical operators in Python. They are "and", "or" and "not". They must be in lowercase.*

# Identitätsoperatoren / *Identity Operators*

Identitätsoperatoren werden verwendet, um Objekte zu vergleichen. Nicht, ob sie gleich sind, sondern ob sie tatsächlich dasselbe Objekt mit derselben Speicheradresse sind.

---

*Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.*

````python
a = [1, 2]
b = [1, 2]
c = a
````

| Operator | Meaning               | Example      | Result  |
| -------- | --------------------- | ------------ | ------- |
| `is`     | Same object in memory | `a is c`     | `True`  |
| `is`     | Same value?           | `a is b`     | `False` |
| `==`     | Same value            | `a == b`     | `True`  |
| `is not` | Not same object       | `a is not b` | `True`  |


Der Unterschied zwischen `is` und `==`:
- `is` – Prüft, ob beide Variablen auf dasselbe Objekt im Speicher verweisen
- `==` – Prüft, ob die Werte beider Variablen gleich sind

---

*The difference Between `is` and* `==`:
- *`is` - Checks if both variables point to the same object in memory*
- `==` *- Checks if the values of both variables are equal*


````python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)
````

````python
value = False

print(value is None)
print(value == None)
````

# Bitweise Operatoren / *Bitwise Operators*

Bitweise Operatoren werden normalerweise verwendet, um bitweise Operationen auf Objekten des Typs Integer durchzuführen. Anstatt das Objekt als Ganzes zu behandeln, wird es als eine Bitfolge betrachtet. Auf jedes Bit in der Folge werden verschiedene Operationen angewendet.

Python verfügt über sechs bitweise Operatoren: &, |, ^, ~, << und >>. Alle diese Operatoren (außer ~) sind binär, das heißt, sie wirken auf zwei Operanden. Jeder Operand ist eine Binärziffer (Bit) mit dem Wert 1 oder 0.

---

*Bitwise operators are normally used to perform bitwise operations on integer-type objects. However, instead of treating the object as a whole, it is treated as a string of bits. Different operations are done on each bit in the string.*

*Python has six bitwise operators - &, |, ^, ~, << and >>. All these operators (except ~) are binary in nature, in the sense they operate on two operands. Each operand is a binary digit (bit) 1 or 0.*

````python
a = 5   # 0101
b = 3   # 0011
````

| Operator | Name        | Example  | Binary Result | Decimal |
| -------- | ----------- | -------- | ------------- | ------- |
| `&`      | AND         | `a & b`  | 0001          | 1       |
| `\|`     | OR          | `a \| b` | 0111          | 7       |
| `^`      | XOR         | `a ^ b`  | 0110          | 6       |
| `~`      | NOT         | `~a`     | flip bits     | -6      |
| `<<`     | Left shift  | `a << 1` | 1010          | 10      |
| `>>`     | Right shift | `a >> 1` | 0010          | 2       |


# Slice-Syntax für Sequenzen / *Slice Operator for Sequences*

Ein Bereich von Zeichen aus einem String oder Werte aus einer Liste können mithilfe der Slice-Syntax zurückgegeben werden. Startindex und den Endindex werden durch einen Doppelpunkt getrennt angegeben, um einen Teil des Strings zurückzugeben. Wenn der Startindex weggelassen wird, beginnt der Bereich beim ersten Element; wenn der Endindex weggelassen wird, erstreckt sich der Bereich bis zum letzten Element.

---

*You can return a range of characters from a string or values from a list by using the slice syntax. Specify the start index and the end index, separated by a colon, to return a part of the string. By leaving out the start index, the range will start at the first item and by leaving out the end index, the range will go to the end.*

````python
statement = "always and forever"

# Get the characters from the start to position 5
print(statement[:5])
print(statement[2:5])

# slice from first
print(statement[1:])
print(statement[:8])

# slice till end
print(statement[5:])
````

````python
# negative slicing
print(statement[-8:-2])
````

````python
# reverse a string
print(statement[::-1])
````
