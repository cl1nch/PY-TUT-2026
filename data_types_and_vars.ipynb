# Datentypen und Variablen / *Data Types and Variables*

Variablen werden verwendet, um Daten zuzuweisen, auf die während der Programmausführung verwiesen und die manipuliert werden können. Es ist im Wesentlichen ein Name, der einem Wert zugewiesen wird. Python hat keinen speziellen Befehl zum Deklarieren einer Variablen. Eine Variable wird in dem Moment erstellt, in dem ihr zum ersten Mal ein Wert zugewiesen wird.

---

*Variables are used to assign data that can be referenced and manipulated during program execution. It is essentially a name that is assigned to a value. Python has no special command for declaring a variable. A variable is created the moment you first assign a value to it.*

```python
# define a variable in Python
x = 5
print(x)
name = "Alice"
home = "wonderland"
print(name, home)
```

Variablen müssen nicht mit einem bestimmten Datentyp deklariert werden, und der Typ kann sogar geändert werden, nachdem sie festgelegt wurden.

---

*Variables do not need to be declared with any particular data type, and the type can even be changed after they have been set.*

```python
x = 4       # x is of type int
print(type(x))
x = "Alice" # x is now of type str
print(type(x))
```

Variablen sind reservierte Speicherorte, die verwendet werden, um Werte innerhalb eines Python-Programms zu speichern. Das bedeutet, dass beim Erstellen einer Variablen ein bestimmter Speicherplatz reserviert wird. Die eingebaute Python-Funktion `id()` gibt die Adresse zurück, an der die Daten gespeichert sind.

---

*Variables are the reserved memory locations used to store values with in a Python Program. This means that when you create a variable you reserve some space in the memory. Python's built-in `id()` function returns the address where the data is stored.*

```python
month = "march"
date = 18

print(id(month))
print(id(date))
```

# Regeln für die Namensgebung / *Rules for Naming Variables*

Jede Variable sollte einen aussagekräftigen und eindeutigen Namen haben, um Lesbarkeit und Wiederverwendbarkeit zu gewährleisten. Beim Benennen einer Python-Variablen sollten bestimmte Regeln beachtet werden:
<ol> <li>Variablennamen dürfen nur Buchstaben, Ziffern und Unterstriche (_) enthalten. Keine Sonderzeichen erlaubt.</li> <li>Ein Variablenname darf nicht mit einer Ziffer beginnen. Er darf nur mit einem Buchstaben oder Unterstrich starten.</li> <li>Variablennamen sind case-sensitiv, z. B. sind `name` und `NaMe` unterschiedlich.</li> <li>Reservierte Schlüsselwörter wie `if`, `else`, `for` usw. dürfen nicht als Variablennamen verwendet werden.</li> </ol>
Wenn der Name der Variablen aus mehreren Wörtern bestehen soll, um ihn aussagekräftiger zu machen, sollten eines der folgenden Benennungsmuster verwendet werden:
<ol> <li>Camel Case: `kmPerHour`, `pricePerLitre`</li> <li>Pascal Case: `KmPerHour`, `PricePerLitre`</li> <li>Snake Case: `km_per_hour`, `price_per_litre`</li> </ol>

---

*Every variable shoould have a meaningful and unique name to ensure readability and reusability. There are certain rules that should be taken into account while naming a python variable:*
<ol>
<li>Variable names can only contain letters, digits and underscores (_). No special characters allowed.</li>
<li>A variable name cannot start with a digit. It can only start with a letter or underscore. </li>
<li>Variable names are case sensitive like name and NaMe are different.</li>
<li>Reserved keywords like if, else, for and so on cannot be used as variable names.</li>
</ol>
    
*If the name of the variable should contain multiple words to make it more meaningful, we should use one of the following naming patterns:*
<ol>
<li>Camel case: kmPerHour, pricePerLitre</li>
<li>Pascal case: KmPerHour, PricePerLitre</li>
<li>Snake Case: km_per_hour, price_per_litre</li>
</ol>

```python
# variables with correct names

_count  = 100
name1 = "Alice"
name2 = "bunny"
Age  = 10
alice_home = "wonderland"

print (_count)
print (name1)
print (name2)
print (Age)
print (alice_home)
```

### Numerische und boolesche Typen / *Numeric and Boolean Type*


Python-Zahlen repräsentieren Daten, die einen numerischen Wert haben. Ein numerischer Wert kann eine ganze Zahl, eine Gleitkommazahl oder sogar eine komplexe Zahl sein.

<ol> <li>Ganze Zahlen (Integers): Positive oder negative ganze Zahlen (ohne Brüche oder Dezimalstellen).</li> <li>Gleitkommazahlen (Float): Eine reelle Zahl mit einer Gleitkommadarstellung. ### Sequenzen / *Sequence Type*

Eine Sequenz ist eine geordnete Sammlung von Elementen, die denselben oder unterschiedliche Datentypen haben können. Sequenzen ermöglichen die Speicherung mehrerer Werte auf eine organisierte und effiziente Weise. In Python gibt es mehrere Sequenzdatentypen.

---

*A sequence is an ordered collection of items, which can be of similar or different data types. Sequences allow storing of multiple values in an organized and efficient fashion. There are several sequence data types of Python.*

#### List

Eine Liste ist einer der eingebauten Datentypen in Python, die verwendet wird, um Sammlungen von Daten zu speichern. Listen ähneln den Arrays, die in anderen Programmiersprachen zu finden sind. Sie sind eine geordnete und veränderbare Sammlung von Elementen. In Python müssen die Elemente einer Liste nicht denselben Datentyp haben.

Listen können erstellt werden, indem eine Sequenz von Elementen einfach in eckige Klammern `[]` gesetzt wird oder indem der Listenkonstruktor `list()` verwendet wird.

---

*List is one of the built-in data types in Python used to store collections of data. Lists are similar to arrays found in other programming languages. They are an ordered and mutable collection of items. In python, items in a list do not need to be of the same data type.*

*Lists can be created by just placing sequence of items inside square brackets `[]` or using a list constructor `list()`.*

```python
# creating a list

fruits = ["apple", "banana", "cherry"]
mixed_fruits = list(("apple", "banana", "cherry", "apple", "cherry"))

print(fruits, type(fruits))
print(mixed_fruits, type(mixed_fruits))
```

```python
# list with mixed data types
person_info = ["Kevin", 34, True, "male"]

print(person_info, type(person_info))
```

Um die Anzahl der Elemente in einer Liste zu erhalten, oft als „Länge der Liste“ bezeichnet, verwenden wir die Funktion `len()`.

---

*To get the number of items in a list, often referred to as "length of a list", by using `len()` function.*

```python
print(len(person_info))
print(len(fruits))
```

Listen sind geordnet, das bedeutet, dass die Elemente in derselben Reihenfolge gespeichert werden, in der wir sie beim Erstellen der Liste hinzugefügt haben. Eine Liste ist jedoch veränderlich, das heißt, wir können Elemente hinzufügen, entfernen oder ändern. Die Elemente sind indexiert: Das erste Element hat den Index 0, das zweite Element den Index 1 usw.

---

*Lists are ordered meaning the items would be stored in the same order we have added them while defining a list. However, a list is changable and we can add, remove or change an item in a list. The items are indexed, the first item has index 0, the second item has index 1 etc.*

```python
# get first item
print("first item", person_info[0])
print("Third item", person_info[2])

# get last item
num_items = len(person_info)
print(person_info[3], person_info[num_items-1], person_info[-1])
```

#### F-Strings

Um einen String als f-String zu kennzeichnen, setzten wir einfach ein f vor den String-Literal und verwendet geschweifte Klammern {} als Platzhalter für Variablen und andere Operationen. Ein Platzhalter kann Variablen, Operationen, Funktionen und Modifikatoren enthalten, um den Wert zu formatieren.

---

*To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations. A placeholder can contain variables, operations, functions, and modifiers to format the value.*

```python
# format string, age is the placeholder
age = 10
txt = f"My name is Alice. I am {age} years old."
print(txt)
```

Ein Platzhalter kann einen Modifikator enthalten, um den Wert zu formatieren.

Ein Modifikator wird hinzugefügt, indem ein Doppelpunkt : gefolgt von einem zulässigen Formatierungstyp angegeben wird, zum Beispiel .2f, was eine Fließkommazahl mit 2 Dezimalstellen bedeutet.

---

*A placeholder can include a modifier to format the value.*

*A modifier is included by adding a colon : followed by a allowed formatting type, like .2f which means fixed point number with 2 decimals:*

```python
pi = 3.1429
txt = f"The value of pi is {pi:.2f}"
print(txt)
```

```python
# place holder with a math operation
txt = f"The value of pi is {22/7}"
print(txt)

txt = f"The value of pi is {22/7:.4f}"
print(txt)
```

#### Tuple

Ein Python-Tuple ist ein weiterer Sequenzdatentyp, der einer Liste ähnelt. Ein Tuple besteht aus einer Anzahl von Werten, die durch Kommas getrennt sind. Im Gegensatz zu Listen werden Tuples jedoch in Klammern (...) eingeschlossen.

Ein Tuple ist ebenfalls eine Sequenz, daher hat jedes Element im Tuple einen Index, der seine Position in der Sammlung angibt. Der Index beginnt bei 0.

Zum Erstellen eines Tuples ist die Verwendung von Klammern optional. Daten, die durch Kommas getrennt sind, ohne dass sie von Klammern eingeschlossen werden, werden standardmäßig als Tuple behandelt.

---

*Python tuple is another sequence data type that is similar to a list. A Python tuple consists of a number of values separated by commas. Unlike lists, however, tuples are enclosed within parentheses `(...)`.*

*A tuple is also a sequence, hence each item in the tuple has an index referring to its position in the collection. The index starts from 0.*

*To form a tuple, use of parentheses is optional. Data items separated by comma without any enclosing symbols are treated as a tuple by default.*

```python
example_tuple = (2023, "Python", 3.11, 5+6j, 1.23E-4)
print(example_tuple, type(example_tuple))
```

```python
example_tuple = tuple((2023, "Python", 3.11, 5+6j, 1.23E-4))
print(example_tuple, type(example_tuple))
```

```python
# an item in the tuple may also be a list
example_tuple = (['One', 'Two', 'Three'], 1,2.0,3, (1.0, 2.0, 3.0))
print(example_tuple, type(example_tuple))
```

```python
# some more sequence style operations
person_tuple = ('Nyx', 'Granger', 34, 'engineer', 'male', '24 Breverly Road')
tinytuple = ('Paradise City', 45609)

print (person_tuple)               # Prints the complete tuple
print (person_tuple[0])            # Prints first element of the tuple
print (person_tuple[1:3])          # Prints elements of the tuple starting from 2nd till 3rd
print (person_tuple[2:])           # Prints elements of the tuple starting from 3rd element
print (tinytuple * 2)       # Prints the contents of the tuple twice
print (person_tuple + tinytuple)   # Prints concatenated tuples
```

Der Hauptunterschied zwischen Listen und Tuples besteht darin, dass Listenelemente und die Größe geändert werden können, d. h. Listen sind veränderlich (mutable), während Tuples nach der Definition nicht geändert werden können (immutable). Tuples können als schreibgeschützte Listen betrachtet werden.

Der folgende Code ist bei Tuples ungültig, da versucht wird, ein Tuple zu ändern, was nicht erlaubt ist.

---

*The main difference between lists and tuples is that list elements and size can be modified i.e. lists are mutable, while tuples cannot be updated (immutable) once defined. Tuples can be thought of as read-only lists.*

*The following code is invalid with tuple, because we attempted to update a tuple, which is not allowed.*

#### Mengen / *Set*

Eine Menge bezeichnet eine Sammlung von eindeutigen Objekten. In Python ist ein Set eine ungeordnete Sammlung von einzigartigen Elementen. Im Gegensatz zu Listen oder Tuples erlauben Sets keine doppelten Werte, d. h., jedes Element in einem Set muss einzigartig sein. Sets sind veränderlich (mutable), das heißt, Sie können Elemente hinzufügen oder entfernen, nachdem ein Set erstellt wurde.

Sets werden mit geschweiften Klammern {} oder der eingebauten Funktion set() definiert. Sie sind besonders nützlich zum Überprüfen von Mitgliedschaften, Entfernen von Duplikaten aus einer Sequenz und für gängige mathematische Mengenoperationen wie Vereinigung, Schnittmenge und Differenz.

---

*A set refers to a collection of distinct objects. In Python, a set is an unordered collection of unique elements. ##### Hinzufügen von Elementen zur Menge / *Adding Elements in a Set*

Um ein Element zu einem Set hinzuzufügen, können wir die Funktion add() verwenden. Dies ist nützlich, wenn wir neue Elemente in ein bestehendes Set aufnehmen möchten. Ist das Element bereits im Set vorhanden, bleibt das Set unverändert.

---

*To add an element to a set, you can use the add() function. This is useful when you want to include new elements into an existing set. If the element is already present in the set, the set remains unchanged.*

```python
# Initialize set with values
graphicDesigner = {'InDesign', 'Photoshop', 'Acrobat', 'Premiere', 'Bridge'}

# add 'Illustrator'
graphicDesigner.add('Illustrator')
print(graphicDesigner)

# add 'Photoshop'
graphicDesigner.add('Photoshop')
print(graphicDesigner)
```

Es ist wichtig zu beachten, dass wir einem Set nur Werte hinzufügen können, die unveränderlich (immutable) sind, wie zum Beispiel ein String oder ein Tuple. Wenn wir beispielsweise versuchen, eine Liste zu einem Set hinzuzufügen, erhalten wir einen TypeError.

---

*It is important to note that you can only add a value that is immutable (like a string or a tuple) to a set. For example, you would get a TypeError if you try to add a list to a set.*

```python
graphicDesigner.add(['Powerpoint', 'Blender'])
```

##### Entfernen von Elementen aus der Menge / *Removing Elements from a Set*

Wir können ein Element aus einem Set mit der Funktion remove() entfernen. Dies ist nützlich, wenn wir bestimmte Elemente aus dem Set löschen möchten. Ist das Element nicht vorhanden, wird ein KeyError ausgelöst.
Alternativ können wir die Funktion discard() verwenden, um ein Element aus dem Set zu entfernen, falls es vorhanden ist. Im Gegensatz zu remove() löst discard() keinen Fehler aus, wenn das Element im Set nicht gefunden wird.

---

*You can remove an element from a set using the remove() function. This is useful when you want to eliminate specific elements from the set. If the element is not present, a KeyError is raised.
Alternatively, you can use the discard() function to remove an element from the set if it is present. Unlike remove(), discard() does not raise an error if the element is not found in the set.*

```python
# Removes the element 'Illustrator' from the set
graphicDesigner.remove('Illustrator')
print(graphicDesigner)
```

```python
# try and remove the element 'Muse' from the set
graphicDesigner.remove('Muse')
print(graphicDesigner)
```

```python
# No error even if 'Muse' is not in the set
graphicDesigner.discard('Muse')
print(graphicDesigner)
```

##### Set Operations

In Python unterstützen Sets verschiedene Mengenoperationen, die zur Manipulation und zum Vergleich von Sets verwendet werden. Diese Operationen umfassen Vereinigung (union), Schnittmenge (intersection), Differenz (difference), symmetrische Differenz (symmetric difference) und Überprüfung auf Teilmengen (subset testing). Sets sind besonders nützlich, wenn wir mit Sammlungen von eindeutigen Elementen arbeiten und Operationen auf Basis der Mengenlehre durchführen möchten.

| Operation              | Symbol / Methode                         | Beschreibung                                                    | Beispiel                  | Ergebnis        |
| ---------------------- | ---------------------------------------- | --------------------------------------------------------------- | ------------------------- | --------------- |
| Vereinigung            | `A \| B` oder `A.union(B)`               | Kombiniert alle Elemente beider Sets                            | `A \| B`                  | `{1,2,3,4,5,6}` |
| Schnittmenge           | `A & B` oder `A.intersection(B)`         | Elemente, die in beiden Sets vorkommen                          | `A & B`                   | `{3,4}`         |
| Differenz              | `A - B` oder `A.difference(B)`           | Elemente in A, die nicht in B sind                              | `A - B`                   | `{1,2}`         |
| Differenz              | `B - A`                                  | Elemente in B, die nicht in A sind                              | `B - A`                   | `{5,6}`         |
| Symmetrische Differenz | `A ^ B` oder `A.symmetric_difference(B)` | Elemente, die in einem der Sets, aber nicht in beiden vorkommen | `A ^ B`                   | `{1,2,5,6}`     |
| Teilmenge              | `A <= B` oder `A.issubset(B)`            | Prüft, ob A in B enthalten ist                                  | `{1,2} <= A`              | `True`          |
| Obermenge              | `A >= B` oder `A.issuperset(B)`          | Prüft, ob A alle Elemente von B enthält                         | `A >= {1,2}`              | `True`          |
| Disjunkt               | `A.isdisjoint(B)`                        | Prüft, ob die Sets keine gemeinsamen Elemente haben             | `{1,2}.isdisjoint({3,4})` | `True`          |


---

In Python, sets support various set operations, which is used to manipulate and compare sets. These operations include union, intersection, difference, symmetric difference, and subset testing. Sets are particularly useful when dealing with collections of unique elements and performing operations based on set theory.

| Operation            | Symbol / Method                        | Description                            | Example                              | Result      |
| -------------------- | -------------------------------------- | -------------------------------------- | ------------------------------------ | ----------- |
| Union                | `A \| B`or`A.union(B)`                       | Combines all elements from both sets | `A \| B` | `{1,2,3,4,5,6}` |
| Intersection         | `A & B` or `A.intersection(B)`         | Elements common to both sets           | `A & B`                              | `{3,4}`     |
| Difference           | `A - B` or `A.difference(B)`           | Elements in A but not in B             | `A - B`                              | `{1,2}`     |
| Difference           | `B - A`                                | Elements in B but not in A             | `B - A`                              | `{5,6}`     |
| Symmetric Difference | `A ^ B` or `A.symmetric_difference(B)` | Elements in either set but not both    | `A ^ B`                              | `{1,2,5,6}` |
| Subset               | `A <= B` or `A.issubset(B)`            | Checks if A is contained in B          | `{1,2} <= A`                         | `True`      |
| Superset             | `A >= B` or `A.issuperset(B)`          | Checks if A contains all elements of B | `A >= {1,2}`                         | `True`      |
| Disjoint             | `A.isdisjoint(B)`                      | Checks if sets have no common elements | `{1,2}.isdisjoint({3,4})`            | `True`      |


```python
dataScientist = set(['Python', 'R', 'SQL', 'Git', 'Tableau', 'SAS'])
dataEngineer = set(['Python', 'Java', 'Scala', 'Git', 'SQL', 'Hadoop'])

print("Union:", dataScientist | dataEngineer)
print("Intersection:", dataScientist & dataEngineer)
print("Difference:", dataScientist - dataEngineer ," erscheint in Science, aber nicht Engineer" )
print("Difference:", dataEngineer - dataScientist)
print("Symmetric Difference:", dataScientist ^ dataEngineer)
```

#### Dictionary

Ein Python-Wörterbuch (Dictionary) ist eine Datenstruktur, die Informationen in Schlüssel-Wert-Paaren speichert. Während Schlüssel eindeutig und unveränderlich (immutable) sein müssen, wie z. B. Strings oder Zahlen, können die Werte beliebige Datentypen haben, sowohl veränderlich als auch unveränderlich, sogar ein weiteres Wörterbuch. Dies macht Dictionaries ideal, um auf Daten über einen bestimmten Namen zuzugreifen, anstatt über eine numerische Position wie bei Listen.
Ein Python-Dictionary ähnelt assoziativen Arrays oder Hashes, wie sie in anderen Programmiersprachen vorkommen, und besteht aus Schlüssel:Wert-Paaren. Die Paare werden durch Kommas getrennt und in geschweifte Klammern {} gesetzt. Um die Zuordnung zwischen Schlüssel und Wert herzustellen, wird zwischen beiden ein Doppelpunkt : gesetzt. Werte für einen bestimmten Schlüssel können jedoch mithilfe von eckigen Klammern [] zugewiesen und abgerufen werden.

---

*A Python dictionary is a data structure that stores information in key-value pairs. While keys must be unique and immutable (like strings or numbers), the values can be of any data type, whether mutable or immutable, even a dictionary. This makes dictionaries ideal for accessing data by a specific name rather than a numeric position like in list.
Python dictionary is like associative arrays or hashes  found in other promgramming languages and consist of key:value pairs. The pairs are separated by comma and put inside curly brackets `{}`. To establish mapping between key and value, the colon `:` symbol is put between the two. However, values for a specific key can be assigned and accessed using square braces (`[]`).*


```python
# define a dictionary
car_dict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(car_dict)
```

```python
# using dict() constructor
car_dict = dict(brand = "Ford", model = "Mustang", year = "1964")
print(car_dict)
```

In Python ist ein Dictionary ein Objekt der eingebauten dict-Klasse. Dies kann mit der Funktion type() überprüft werden.

---

*In Python, dictionary is an object of the built-in dict class. We can check it with the type() function.*

```python
print(type(car_dict))
```

##### Zugriff auf Dictionary-Elemente / *Accessing Dictionary Items*

Ein Wert in einem Dictionary wird über seinen Schlüssel wie über einen Index abgerufen. Dies kann entweder mit eckigen Klammern [ ] oder mit der Methode get() erfolgen. Beide geben den Wert zurück, der dem angegebenen Schlüssel zugeordnet ist.

---

*A value in a dictionary is accessed by using its key like an index. This can be done either with square brackets `[ ]` or with the `get()` method. Both return the value linked to the given key.*

```python
# accessing item using just key as index
print(car_dict['brand'])

# accessing item using get
print(car_dict.get('model'))
```

```python
# try to access a key that does not exist
print(car_dict.get('license'))
```

```python
# access a value if key exists, else return an empty string
print(car_dict.get('license', 'license not found'))
```

```python
# items
print("items: ", car_dict.items())
# keys
print("keys: ", car_dict.keys())
# values
print("values: ", car_dict.values())
```

Dictionaries sind veränderlich (changeable), das heißt, wir können Elemente ändern, hinzufügen oder entfernen, nachdem das Dictionary erstellt wurde. Neue Elemente werden einem Dictionary mit dem Zuweisungsoperator = hinzugefügt, indem einem neuen Schlüssel ein Wert zugewiesen wird. Wird ein vorhandener Schlüssel mit dem Zuweisungsoperator verwendet, wird sein Wert durch den neuen ersetzt. Duplikate sind jedoch nicht erlaubt; ein Dictionary kann nicht zwei Elemente mit demselben Schlüssel haben.

Ein Python-Dictionary ist keine Sequenz. Es ist eine Sammlung von Elementen, aber jedes Element (Schlüssel:Wert-Paar) wird nicht durch einen Positionsindex wie bei String, Liste oder Tuple identifiziert. Daher kann auf einem Dictionary keine Slicing-Operation durchgeführt werden.

---

*Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created. New items are added to a dictionary using the assignment operator (=) by giving a new key a value. If an existing key is used with the assignment operator, its value is updated with the new one. However, duplicates are not allowed. Dictionaries cannot have two items with the same key.*

*Python's dictionary is not a sequence. It is a collection of items but each item (key:value pair) is not identified by positional index as in string, list or tuple. Hence, slicing operation cannot be done on a dictionary.*

```python
# change the car model
car_dict["model"] = "fiesta"
print(car_dict)
```

```python
# add car license plate
car_dict["license"] = "MN1337"
print(car_dict)
```

```python
# try adding a duplicate key
car_dict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020,
  "license": "MN1337"
}
print(car_dict)
```

##### Entfernen von Dictionary-Elementen / *Removing Dictionary Items*

Dictionary-Elemente können mit eingebauten Löschmethoden entfernt werden, die auf Schlüssel basieren:
* del: entfernt ein Element über seinen Schlüssel
* pop(): entfernt das Element mit dem angegebenen Schlüssel und gibt seinen Wert zurück
* clear(): entfernt alle Elemente aus dem Dictionary
* popitem(): entfernt und gibt das zuletzt eingefügte Schlüssel-Wert-Paar zurück

---

*Dictionary items can be removed using built-in deletion methods that work on keys:*

- `del`: *removes an item using its key*
- `pop()`: *removes the item with the given key and returns its value*
- `clear()`: *removes all items from the dictionary*
- `popitem()`: *removes and returns the last inserted key–value pair*

```python
# delete license
del car_dict['license']
print(car_dict)
```

```python
# Using pop() to remove model
val = car_dict.pop('model')
print(val)
print(car_dict)

# Using popitem()
key, val = car_dict.popitem()
print(f"Key: {key}, Value: {val}")
print(car_dict)

# Using clear()
car_dict.clear()
print(car_dict)
```
