# The Basics of Scientific Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Einfache Diagramme / *Simple plots*

Lass uns damit beginnen, einfache Kurven basierend auf numerischen Daten zu erstellen, um einige der Möglichkeiten zu verstehen, die wir in der Visualisierung haben.

---

*Let us start by creating simple curves based on numerical data to understand some of the choices we have in visualization.*

```python
data = [1.3, 2.7, 4.2, 6.0, 9.0, 9.75, 10.0, 10.1]
```

```python
fig = plt.figure()
plt.plot(data)
plt.show()
```

Hier können wir sehen, dass der Befehl `plt.figure()` die Figur erstellt, `.plot()` die "Leinwand" mit Daten füllt und der Befehl `.show()` sie anzeigt.

Natürlich ist es auch möglich, mehr als eine Kurve zu zeichnen und dafür NumPy-Arrays zu verwenden. Lasst uns dies ausprobieren, indem wir einige zufällig erzeugte ganze Zahlen visualisieren.

---

*Here, we can see that the `plt.figure()` command creates the figure, the `.plot()` fills the "canvas" with data, and the `.show()` command displays it.*

*Of course, it is possible, to also plot more than one curve and make use of numpy arrays to this end. Let's try this by visualizing some randomly created integers.*

```python
random_data = np.array([np.random.randint(0,10,20), np.random.randint(0,10,20)])
random_data
```

```python
fig = plt.figure()
plt.plot(random_data[0])
plt.plot(random_data[1])
plt.show()
```

# Anpassen von Diagrammen / *Customizing Plots*

In der Praxis sind keine dieser Darstellungen besonders anschaulich.

Einerseits haben wir angenommen, dass die Zahlen durch eine Linie verbunden werden sollten. Mit anderen Worten: Der Wert zwischen zwei aufeinanderfolgenden Einträgen des Arrays wird interpoliert. Dies kann irreführende Zusammenhänge suggerieren. Im ersten Diagramm war diese Wahl möglicherweise besser als im Zweiten.

Andererseits enthalten keine dieser Figuren Beschreibungen darüber, was das Diagramm zeigt, was die x und y Achse darstellen oder welche Bedeutung die Farbe hat. Beachte, dass die Farbe automatisch unterschiedlich gesetzt wurde, wenn mehr als eine Linie dargestellt wird.

Versuchen wir nun, dem Diagramm einen Titel, eine Legende und Achsenbeschriftungen hinzuzufügen.

---

*In practice, neither of these plots are great to look at.*

*On the one hand, we made the assumption that the numbers we have should be connected in a line. In other words, the value in between two subsequent entries of the array is interpolated. This can suggest misleading relationships. In the first plot the choice may have been better than in our second one.*

*On the other hand, neither of these figures contains descriptions of what the plot is showing, what the individual x-axis and y-axis are, or explain the meaning of the color. Note, as you can observe, that the color has been set automatically to a different one for more than one line.*

*Let's try adding a title, legend, and axis labels to the plot.*

```python
fig = plt.figure()
plt.plot(random_data[0])
plt.plot(random_data[1])
plt.xlabel("This is our x-axis")
plt.ylabel("This is our y-axis")
plt.legend(loc=1, labels=["First line", "Second line"])
# Loc specifies the respective part in which the legend is placed
# check the function documentation to see how locations like "upper_right"
# are mapped to specific numbers
plt.title("This is our title")
plt.show()
```

Das ist bereits besser als zuvor, aber wir stellen fest, dass die Schriftgrößen automatisch nach internen Standardwerten gesetzt wurden.

Oft möchten wir diese selbst festlegen, um den allgemeinen Anforderungen eines Dokuments zu entsprechen, die Lesbarkeit zu gewährleisten oder bestimmte Teile besonders hervorzuheben.

---

*This is already better than what we had before, but we note that the font sizes were set automatically according to internal default values.*

*Often times, you may want to specify these yourself, to comply to general requirements of a document you are writing, to ensure legibility, or to emphasize some part very strongly.*

```python
fig = plt.figure()
plt.plot(random_data[0])
plt.plot(random_data[1])
plt.xlabel("This is our x-axis", fontsize=18)
plt.ylabel("This is our y-axis", fontsize=18)
plt.legend(loc=1, labels=["First line", "Second line"], fontsize=12)
plt.title("This is our title", fontsize=24)
plt.show()
```

Da es mühsam und repetitiv ist, dies für viele Diagramme konsequent zu wiederholen, erlaubt matplotlib die Verwendung von Dictionaries.

---

*Because it is tedious and repeptitive to do this consistently for many plots, matplotlib allows you to use dictionaries to this end.*

```python
font_title = {'family': 'serif',
              'color':  'darkblue',
              'weight': 'bold',
              'size': 24}

font_axes = {'family': 'serif',
              'color':  'darkred',
              'weight': 'normal',
              'size': 16}


fig = plt.figure()
plt.plot(random_data[0])
plt.plot(random_data[1])
plt.xlabel("This is our x-axis", fontdict=font_axes)
plt.ylabel("This is our y-axis", fontdict=font_axes)
plt.legend(loc=1, labels=["First line", "Second line"])
plt.title("This is our title", fontdict=font_title)
plt.show()
```

Schließlich können wir die Farbe der Linien selbst ändern.
Dafür gibt es die Option `color`.

---

*Finally, we can modify the color of the lines themselves.*
*For this purpose, the `color` option exists.*

```python
fig = plt.figure()
plt.plot(random_data[0], color='#0f0f0f')
plt.plot(random_data[1], color="green")
plt.xlabel("This is our x-axis", fontdict=font_axes)
plt.ylabel("This is our y-axis", fontdict=font_axes)
plt.legend(loc=1, labels=["First line", "Second line"])
plt.title("This is our title", fontdict=font_title)
plt.show()
```

Beachte, dass es viele Möglichkeiten gibt, Farben anzugeben. Siehe https://matplotlib.org/stable/users/explain/colors/colors.html für eine umfassende Liste. Dies umfasst Strings und Hex-RGB/RGBA-Strings, wie oben gezeigt, Abkürzungen wie 'g' für Grün und vieles mehr.

Da viele Menschen viel Zeit damit verbracht haben, gute Farbskalen für bestimmte Zwecke zu entwickeln, kann es oft sinnvoll sein, eine auszuwählen, die am Besten zu unseren Bedürfnissen passt: https://matplotlib.org/stable/users/explain/colors/colormaps.html
Wir können auf die ListedColormaps von matplotlib über ihr Farb-Attribut zugreifen.

---

*Note that there are many ways to specify colors, see https://matplotlib.org/stable/users/explain/colors/colors.html for a comprehensive list. This includes a string and hex RBG/RBGA strings, as shown above, abbreviations like 'g' for green, and many more.*

*As many people have spent substantial amount of time thinking about good colormaps for particular purposes, it can often be smart to pick one that best fits our needs: https://matplotlib.org/stable/users/explain/colors/colormaps.html
We can index matplotlis ListedColormaps through their color attribute.*

```python
from matplotlib import colormaps
colors = colormaps['Dark2'].colors

fig = plt.figure()
for i, color in enumerate(colors):
    plt.plot([0, i], color=color)
```

Sei jedoch vorsichtig: Wenn die Farben in der Palette "ausgehen", stürzt entweder der obige Code ab, weil der Index außerhalb des zulässigen Bereichs liegt, oder, wenn die Farbpalette im Voraus festgelegt wurde, werden die Farben meist einfach wiederholt.

Zum Beispiel definiert die Standard-Farbpalette nur 10 Farben, sodass die Farben nach der zehnten Linie einfach wiederholt werden.

---

*But be aware, that if you run out of colors, either above code will crash because the index runs out of bounds, or alternatively, if you have set the colormap in advances, then the colors will simply be repeated.*

*For instance, the default color map only defines 10 colors, and as such, colors will simply be repeated after the 10th line has been plotted.*

```python
fig = plt.figure()
for i in range(50):
    plt.plot([0, i])
```

Als letzten Schritt ist es daher oft üblich, in Visualisierungen bestimmte Marker oder Linienstile festzulegen (auch weil sie im Druck besser lesbar sind) und zusätzlich die Linienstärke zu ändern.

---

*As a lest step, it is thus often also common to set specific markers or line styles in visualizations (and also because they will be more legigible in print) and additionally modify line thickness.*

```python
fig = plt.figure()

linestyles = ['-', '--', '-.', ':']

for i in range(4):
    plt.plot([0, i], linestyle=linestyles[i], linewidth=3)
```

Stelle insgesamt sicher, dass:
* alle Diagramme in einem Bericht oder Dokument einen konsistenten Stil haben
* alle Texte groß genug sind
* ein Titel oder eine Beschriftung enthalten ist
* die Achsen beschriftet sind
* die Farben auch für farbenblinde Personen geeignet sind
* farbige Diagramme auch in Graustufen verständlich sind, falls ein Ausdruck vorgesehen ist
* keine Farbskalen wie „jet“ für kontinuierliche Daten verwendet werden, da sie künstliche Grenzen erzeugen, wo keine existieren. Siehe https://www.youtube.com/watch?v=xAoljeRJ3lU für ein Video, warum matplotlib daher seine Standard-Farbkarte geändert hat.

---

*Overall, make sure that*
* *all of your plots have consistent style in a report paper/document*
* *all text is large enough*
* *contain a title or a caption*
* *axes are labeled*
* *colors are suited also for colorblind people*
* *colored plots can still be understood if printed in grayscale, if print is desired*
* *do not use colormaps such as "jet" for continuous data as they will introduce artificial boundaries where there are none. See https://www.youtube.com/watch?v=xAoljeRJ3lU for a video of why matplotlib has therefore changed its default colormap.*
