# Bedingungen / *Conditional Statements*

Bedingungen in Python erlauben es einen bestimmten Block an Code folgend einer Vorraussetzung auszuführen. Mit diesen Statements können wir die Ausführung des Programmes gezielt kontrollieren und dabei verschiedene Situationen abdecken.

---

*Conditional statements in Python are used to execute certain blocks of code based on specific conditions. These statements help control the flow of a program, making it behave differently in different situations.*

#### If Conditional Statement
Das if Statement ist eine simple Bedingung. Ein Block an Code wird ausgeführt, wenn die Bedingung erfüllt ist.
Das folgende Beispiel überprüft, ob ein User wahlberechtigt ist auf Basis des Alters.

---

*If statement is the simplest form of a conditional statement. It executes a block of code if the given condition is true.
The following example checks if the user is eligible to vote based on their age.*

````python
age = 20
if age >= 18:
    # Indentation defines block
    print("Eligible to vote.")
````

##### If else Conditional Statement
If Else erlaubt es uns weiterhin zusätzlich zu spezifizieren, welcher Code ausgeführt werden soll wenn die Bedingung nicht erfüllt ist. Das heißt wenn die if Bedingung als False evaluiert wird. In unserem obigen Beispiel haben wir einen print, wenn ein User alt genug zum Wählen ist. Allerdings wirds nicht ausgegeben, falls diese Bedingung nicht erfüllt ist. Dies können wir nun ändern, indem wir ein `else` statement hinzufügen.

---

*If Else allows us to specify a block of code that will execute if the condition(s) associated with an if or elif statement evaluates to False. The Else block provides a way to handle all other cases that don't meet the specified conditions. Consider the previous example. We printed a statement indicating if a user was eligible to vote. But nothing was printed if they were not. We can remedy that by using an `else` statement.*

````python
age = 20
if age >= 18:
    print("Eligible to vote.")
else:
    print("Not eligible to vote yet.")
````

Notiz: wir können auch eine Bedingung wie `0 <= i <= n` überprüfen und testen, ob der Wert von `i` innerhalb des Bereiches von `0` nach `n` liegt.

---

*Note: we can also check a condition like `0 <= i <= n` to see whether the value of `i` lies within the inclusive range from `0` to `n`.*

##### Ternary Conditional Statement
Das ternary conditional statement (ternäre Bedingung) ist im Grunde eine Kurzform des if-else statements. Dieses erlaubt uns eine kompaktere Schreibweise und kann dadurch die Lesbarkeit fördern.

---

*The ternary conditional statement is essentially a shorthand for the if-else statement that allows us to write more compact and readable code, especially for simple conditions. It is sometimes called a "conditional expression."*

````python
marks = 42
res = "Pass" if marks >= 40 else "Fail"
print(f"Result: {res}")
````

##### elif Statement
Elif steht in Python for "else if". Es ermöglicht uns mehrere Bedingungen zu überprüfen. Dadurch können wir verschiedene Codeblöcke ausführen, je nachdem welche Bedingung erfüllt ist. Elif statements zu benutzen macht unseren Code leserlicher, denn es wird verhindert, dass mehrere unterschiedliche if Statements konstruiert werden müssen. Es ist wichtig zu verstehen, dass die Bedingungen von oben nach unten in der Reihenfolge getestet werden. Die erste Bedingung die erfüllt ist, wird dann ausgeführt.

Stellen wir uns vor, wir wollen z.B. ein Notensystem programmieren, dass auf Basis von Prozentzahlen Noten ausgibt. Dies könnte wie folgt aussehen:

---

*The elif statement in Python stands for "else if". It allows us to check multiple conditions, providing a way to execute different blocks of code based on which condition is true. Using elif statements makes our code more readable and efficient by eliminating the need for multiple nested if statements.
Note that conditions are checked top-down and only first matching block runs.*

*Suppose, we want to build a grading system for school. Our system looks like the following:*

- $85-100$ => Execellent
- $70-85$ => Good
- $50-70$ => Pass
- $< 50$ => Fail
We cannot achieve this by `if-else` alone.

````python
grade = 75

if grade >= 85:
    print("Excellent")
elif grade >= 70:
    print("Good")
elif grade >= 50:
    print("Pass")
else:
    print("Fail")
````

# Schleifen / *Loops*

Es gibt Situationen, in denen Blöcke von Code mehrfach ausgeführt werden sollten.
Schleifen erlauben uns solch eine mehrfache Ausführung.

---

*There may be a situation when you need to execute a block of code several number of times. Loops allow us to execute a statement or group of statements multiple times.*

### For Schleifen / *For Loops*
For Schleifen werden genutzt, um über eine Sequenz zu iterieren, z.B. eine Liste, ein Tuple, Strings oder range. For Schleifen erlauben es uns den Code für jedes Element in der Sequenz auszuführen. Dabei wird in jedem Schritt der gleiche Code ausgeführt. Die For Schleife startet mit for als Keyword, gefolgt von einer Variable, die das aktuelle Element der Sequenz bestimmt. Ein Doppelpunkt (:) wird benutzt um die Schleife zu starten, und ein eingerückter Block an Code wird innerhalb der Schleife dann mehrfach ausgeführt.

---

*For loops is used to iterate over a sequence such as a list, tuple, string or range. It allow to execute a block of code repeatedly, once for each item in the sequence. It performs the same action on each item of the sequence. This loop starts with the for keyword, followed by a variable that represents the current item in the sequence. A colon (:) is used at the end of the loop header, and the indented block of code beneath it is executed once for each item in the sequence.*

````python
n = 4
for idx in range(n): # runs from 0,1,..., n-1
    print(idx)
````

````python
n = 10

for idx in range(2, n, 2):
    print(idx)
````

````python
n = 10
# range(start, stop, increment)
for idx in range(1, n, 2):
    print(idx)
````

# Inline Schleifen / *Inline For Loop (List Comprehension)*

Eine inline For Schleife in Python wird oft auch als list comprehension bezeichnet.
Sie erlaubt uns kompakten code zu schreiben. Es wird eine neue Liste erzeugt, wobei der Ausdruck für jedes Element der Liste angewandt wird.

---

*An inline for loop in Python, also called a list comprehension, is a concise way to create a new list by applying an expression to each element in an existing sequence.*

````python
numbers = [1, 2, 3, 4]
squares = [x**2 for x in numbers]

print(squares)
````

````python
numbers = [1, 2, 3, 4, 5]
even = [x for x in numbers if x % 2 == 0]

print(even)
````

# While Schleifen / *While Loops*

Eine While Schleife führt wiederholt ein Fragment von Code aus, solange eine Bedingung erfüllt bleibt (als True evaluiert wird).
The Schleife startet mit `while` als keyword, gefolgt von einer boolschen Bedingung und einem Doppelpunkt (:). Diesem folgt dann wieder ein eingerücktes Stück Code zum ausführen.

Der eingerückte Code kann nur eine einzelne Zeile enthalten, kann aber auch mehrzeilig sein und aus mehreren Statements bestehen die der gleichen Einrückung folgen. Sobald die Bedingung zu False evaluiert wird, wird die Schleife beendet und das Programm springt zur nächsten nicht-eingerückten Codezeile nach der Schleife.

```
while expression:
   statement(s)
```

---

*A while loop in Python programming language repeatedly executes a target statement as long as a specified boolean expression is true. This loop starts with `while` keyword followed by a boolean expression and colon symbol (:), followed by an indented block of statements.*

*There may be a single statement or a block of statements with uniform indent. As soon as the conditional expression becomes false, the loop terminates and program control passes to the line immediately following the loop.*

```
while expression:
   statement(s)
```

````python
# a simple while loop
count = 0
while count < 5:
   count += 1
   print (f"Iteration no. {count}")

print ("End of while loop")
````

# Loop Control Statements

Loop control statements erlauben es uns die Logik einer Schleife (for oder while) weiter zu kontrollieren.
Durch sie können wir das verhalten modifizieren, z.B. früher stoppen, eine Iteration überspringen, oder temporär passen.
Python erlaubt die folgenden control statements:
- `break`
- `continue`
- `pass`

---

*Loop control statements in Python are special statements that help control the execution of loops (for or while). They let you modify the default behavior of the loop, such as stopping it early, skipping an iteration, or doing nothing temporarily. Python supports the following control statements:*
- `break`
- `continue`
- `pass`

### Break Statement

Das `break` Statement wird benutzt um eine Schleife zu verlassen, ein frühzeitiger "break", noch bevor die eigentliche Iteration vollständig abgeschlossen ist oder die Terminalbedingung erfüllt ist. Wenn die break Bedingung erfüllt ist, springt das Programm direkt weiter zum Code hinter der Schleife.

---

*The `break` statement in Python is used to exit or “break” out of a loop (either a for or while loop) prematurely, before the loop has iterated through all its items or reached its terminating condition. When the break statement is executed, the program immediately exits the loop, and the control moves to the next line of code after the loop.*

````python
# Using For Loop
for i in range(5):
    if i == 3:
        break  # Exit the loop when i is 3
    print(i)

# Using While Loop
i = 0
while i < 5:
    if i == 3:
        break  # Exit the loop when i is 3
    print(i)
    i += 1
````

### Continue Statement

In Python zwingt das `continue` statement den nächsten Schritt der Iteration innerhalb einer Schleife direkt auszuführen. Als solches wird jeglicher code innerhalb der Schleife unterhalb des continue statements überspringen und erst wieder in der nächsten Iteration ausgeführt (falls hier nicht auch ein continue enthalten ist).

---

*In Python, `continue` forces to execute the next iteration of the loop while skipping the rest of the code inside the loop for the current iteration only, i.e. when the continue statement is executed in the loop, the code inside the loop following the continue statement will be skipped for the current iteration and the next iteration of the loop will begin.*

````python
for i in range(5):
    if i == 3:
        continue  # Skip the rest of the code for i = 3
    print(i)
````

### Pass Statement

Das `pass` statement ist ein Platzhalter, bzw. eine Operation die weiter nichts macht.
Es wird oft benutzt um sicherzustellen, dass das Programm syntaktisch korrekt ist und daher ausführbar bleibt.
Es hat keine weiteren Folgen, außer die Struktur des Programms zu bewahren.

---

*The `pass` statement in Python is a null operation or a placeholder. It is used when a statement is syntactically required but we don’t want to execute any code. It does nothing but allows us to maintain the structure of our program.*

````python
for i in range(5):
    if i == 3:
        pass  # Placeholder for future code
    print(i)
````
