### Installieren von Paketen / *How to install packages*

Wie zu erwarten, erlaubt `install` die Installation von Paketen, wobei das Paket standardmäßig aus dem Python Package Index bezogen wird, falls es dort existiert.

---

*Unsurprisingly, `install` allows you to install packages, where the package will be taken from the Python Package Index by default, if it exists.*

````python
!python -m pip install sampleproject
````

Alternativ können wir URLs angeben, zum Beispiel von GitHub, oder auf Dateien verweisen, zum Beispiel im `.tar.gz`-Format. Die Option `--upgrade` aktualisiert ein Paket, anstatt es zu installieren, vorausgesetzt, es existiert bereits. Der Befehl `uninstall` deinstalliert entsprechend ein Paket.

Es ist üblich, dass größere Projekte viele verschiedene Bibliotheken verwenden. Daher kann es sinnvoll sein, alle Paketabhängigkeiten in einer sogenannten `requirements.txt`-Datei anzugeben. Hier können alle Pakete aufgelistet werden, einschließlich ihrer genau benötigten Versionen, falls erforderlich (wenn nicht angegeben, wird die neueste Version verwendet), und alle Pakete können rekursiv mit folgendem Befehl installiert werden:

```
python -m pip install -r requirements.txt
```

---

*As an alternative, we can specify URLs, for instance from GitHub, or point to files, for instance in `.tar.gz` format. The `--upgrade` option will upgrade a package, rather than installing it, assuming it already exists. The `uninstall` command will respectively uninstall a package.*

*It is common for larger projects to make use of many different libraries. As such, it can be recommended to specify all package dependencies in a so called `requirements.txt` file. Here, we can list all packages, including their exact required versions if needed (if left unspecified, the most recent version will be taken) and recursively install all of them with the following command.*

```
python -m pip install -r requirements.txt
```

Schließlich kann es sein, dass uns Systemberechtigungen fehlen oder, dass wir die globale Python-Installation, die alle Benutzer des Systems betrifft, nicht ändern wollen. Wir können die Option `--user` zu unserem Installationsbefehl hinzufügen, um sicherzustellen, dass die installierten Pakete nur für einen Benutzer verfügbar sind und daher keine sudo-Berechtigungen erfordern. Colab ist standardmäßig bereits eine lokale Instanz, sodass dies keinen Unterschied macht, aber wir können den Befehl trotzdem erfolgreich ausführen.

Die Option `--user` ist nützlich, löst jedoch keine Konflikte zwischen Paketen und Versionen, wenn verschiedene Projekte parallel verwalten werden sollen. Zu diesem Zweck gibt es ein komplexeres Tooling, das wir im Folgenden erkunden werden.

---


*Finally, you may lack system permissions or would prefer to not touch the global python installation that affects every system user. You can add the `--user` flag to your install command to ensure the installed packages do not extent beyond your user, and as such also do not require sudo permissions. Colab is already by default a local instance, so it won't make a different, but we can execute the command successfully nonetheless.*

*The `--user` flag is useful, but it doesn't resolve conflicts between packages and versions when you want to manage different projects in parallel. For this purpose, a more intricate tooling exists, which we will explore in the following*

### Virtuelle Umgebungen in Python / *Virtual environments in Python*

Wir könnten uns jetzt fragen, wie wir verschiedene Projekte parallel verwalten können oder wie wir die Reproduzierbarkeit unseres Codes sicherstellen, wenn Pakete installiert werden und möglicherweise miteinander in Konflikt geraten.

Während es komplexere Optionen wie Docker gibt (die ebenfalls Requirements-Dateien verwenden), gibt es eine einfache und native Möglichkeit, dies auf unserem eigenen System zu handhaben, die „Virtuelle Umgebung“. Diese wird auch kurz `venv` genannt wird.

Virtuelle Umgebungen erlauben es, Pakete innerhalb einer abgeschotteten Umgebung zu installieren, ohne die globale System- oder benutzerweite Python-Installation zu beeinflussen. Im Wesentlichen muss ein venv erstellt und aktiviert werden, und jedes installierte Paket ist dann lokal enthalten.

Wir haben keine Möglichkeit, dies innerhalb unserer Notebooks zu testen, da Colab bereits unabhängige Instanzen erstellt. Aber wir nutzen dies als Gelegenheit, um in lokale Skripterstellung und das Einrichten der eigenen Umgebung einzutauchen, bevor wir für den Rest des Kurses zu Colab zurückkehren.

Um eine virtuelle Umgebung zu erstellen und zu aktivieren, können wir folgenden Befehl verwenden:

```
python -m venv tutorial_env
source tutorial_env/bin/activate
```

Für grundlegende Fragen zum Einrichten verweisen wir auch auf das offizielle Tutorial: https://packaging.python.org/en/latest/tutorials/installing-packages/

---

*You may now be asking yourself how to manage different projects in parallel, or how to ensure reproducibility of your code, if packages get installed and can potentially conflict with one another.*

*Whereas more complex options like Docker exist (that also makes use of requirements files), there is an easy and native way to deal with this on your system called a "Virtual Environment", or `venv` for short.*

*Virtual environments allow you to install packages within a closed off environment, and without affecting your global system or user-wise python installation. In essence a venv needs to be created and activated, and any installed package will then be locally contained.*

*We don't have a way to test this within our notebooks, as colab itself already creates mutually independent instances. But we will use this as an opportunity to dive into local scripting and setting up your own environment, before returning to Colab for the remainder of the course.*

*To create a virtual environment and activate it, we can use the following command:*

```
python -m venv tutorial_env
source tutorial_env/bin/activate
```

*but we also point to the regular tutorial for set-up questions:* https://packaging.python.org/en/latest/tutorials/installing-packages/


# Scripting & Argparse



````python
import argparse

parser = argparse.ArgumentParser(
    prog='ProgramName',
    description='Describe your program here',
    epilog='if necessary, some text at the end'
    )

parser.add_argument('-c', '--count', help='Specify an integer', default=5, type=int)      # option that takes a value
parser.add_argument('-v', '--verbose', help='Enables verbose printing',
                    action='store_true', default=False)  # on/off flag
parser.add_argument('-f', '--file', type=argparse.FileType('r'),
                    help='Path to file')  # specifies a file
````

Lasst uns uns das Obige aufschlüsseln.
Zunächst können wir den Befehlsparser ohne Angaben aufrufen, sofern wir Standardwerte definiert haben. Diese Werte werden übernommen, wenn nichts anderes angegeben wird.

---

*Let's unpack the above.*
*First of all, we can call the command parser without anything specified, given that we have defined default values. These values will be adopted if nothing else is specified.*

````python
args = parser.parse_args([])
print(args.count)
print(args.verbose)
print(args.file)
````

Der Argumentparser erwartet im Allgemeinen eine Liste, wenn dieser in einem Notebook verwendet wird, zum Beispiel in Colab. Alternativ würde auf der Kommandozeile das `--option`-Flag einfach in der Zeile angegeben, in der wir den Code mit Python aufrufen.

Wir können diese Flags verwenden, um einen anderen Wert für z. B. unsere count-Option festzulegen, für die ein ganzzahliger Wert angegeben werden muss.

---

*The argument parser generally expects a list when used in a Notebook, such as within Colab, or alternatively in the command line the `--option` flag would just be provided in the line where we call the code with python.*

*We can use these flags to set a different value of e.g. our count option that requires an integer type to be specified.*

````python
args = parser.parse_args(['--count', '10'])
print(args.count) ## our default value is overwritten by the specified number
````

Wir können auch die Kurzschreibweise mit einem einzelnen Bindestrich verwenden, was zum gleichen Ergebnis führt. Beachte, dass dies nützlich ist, um die Zeile nicht zu lang werden zu lassen. Es wird jedoch empfohlen, sowohl eine kurze als auch eine lange Variante anzugeben, um Verwirrung zu vermeiden.

---

*We can also use the single dash shorthand notation, which will result in the same output. Note that this is useful to avoid the line becoming overly long, but it is recommended to specify both a short and long variant to avoid confusion.*

````python
args = parser.parse_args(['-c', '10'])
print(args.count) ## our default value is overwritten by the specified number
````

Bei Booleschen Optionen ist es üblich, `action='store_true'` zu verwenden, wodurch der Standardwert False auf True gesetzt wird, sobald das Flag aktiviert wird, ohne dass darüber hinaus etwas angegeben werden muss.

---

*For boolean options it is commong to use `action='store_true'`, which results in the default of False being set to True as seen as the flag is set, without needing to specify anything beyond it.*

````python
args = parser.parse_args(['--count', '10', '--verbose'])
print(args.count)
print(args.verbose)
````

Kommandozeilen-Argumentparser können sehr nützlich sein, wenn wir denselben Code mit unterschiedlichen Optionen ausführen möchten, zum Beispiel in einem wissenschaftlichen Experiment, ohne den Code selbst ändern zu müssen.

In unserem Beispiel könnte das count-Flag eine Zahl sein, die als Eingabe an eine Funktion übergeben wird, die berechnet werden soll, oder die verbose-Option könnte einen Debugging-Modus aktivieren, der einige zusätzliche print-Anweisungen enthält.

Wenn unser Argumentparser wächst, wird es immer weniger offensichtlich, welche Optionen es gibt. Um dem entgegenzuwirken, ist es daher üblich, Hilfebefehle einzufügen, wie wir es bereits oben getan haben. `help` nimmt Strings entgegen, die intuitive Anweisungen zur Nutzung liefern sollen. Wenn wir `parser.print_help()` aufrufen, wird in der Kommandozeile (oder hier in unserem Notebook) eine vollständige Liste der Optionen angezeigt, sodass wir uns bewusst werden können, welche Optionen alles eingestellt werden können.

---

*(Command line) argument parsers can be very useful when we want to execute the same code with different options, for example in a scientific experiment, but without needing to touch the code itself anymore.*

*In our example, the count flag could be a number specified as input to a function to be computed, or the verbose option could enable a debugging mode that includes a couple of print statements.*

*As your argument parser grows, it will be come less and less evident what options all exist. To provide a remedy, it is thus common to include help commands, as we have already done above. `help` takes strings that should provide intuitive use instructions. When we call `parser.print_help()`, a full list of options will be given to the user in the command line (or here, our notebook), so that can become aware of what there all is to set.*

````python
parser.print_help()
````
